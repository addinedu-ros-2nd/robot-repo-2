```python
file_path = '/home/kimdu/hanse/src/ai_manipulation/ai_manipulation/script/log.txt'  # 파일 경로

with open(file_path, 'r') as file:
    lines = file.readlines()  # 파일의 각 라인을 리스트로 읽음

with open(file_path, 'w') as file:
    for line in lines:
        if 'episode succeed' in line:
            words = line.split()
            for i, word in enumerate(words):
                if 'episode' in word:
                    times_number = int(''.join(filter(str.isdigit, words[i+2])))  # times 앞의 숫자 추출

                    if times_number <= 10:
                        file.write(line)  # times 앞의 숫자가 10 이하인 문장은 그대로 씀
                    break  # episode succeed가 발견되면 루프 종료

```

```python
from collections import Counter
import json

# 예제 리스트
file_path = "/home/kimdu/hanse/src/ai_manipulation/ai_manipulation/datas/way_point_2.json"
with open(file_path, 'r') as json_file:
    my_list1 = json.load(json_file)
file_path = "/home/kimdu/hanse/src/ai_manipulation/ai_manipulation/datas/six_dxl_acts.json"
with open(file_path, 'r') as json_file:
    my_list2 = json.load(json_file)

counter1 = Counter(str(item[:5]) for item in my_list1)
counter2 = Counter(str(item[:5]) for item in my_list2)
counter1.update(counter2)

# 함수 호출
idx = 1
learning_again = []
for item, count in counter1.items():
    print(f"{idx}. {item}: {count-1}")
    idx += 1
if count < 10:
    print("learn again :", f"{idx}")
    learning_again.append(idx)
```

```python
from collections import Counter
import json

def count_duplicates(lst):
    # 리스트 내의 항목별 중복 횟수를 계산
    counts = Counter(str(item) for item in lst)
    # 결과를 출력
    idx = 1
    for item, count in counts.items():
        print(f"{idx}. {item}: {count}")
        idx += 1
# 예제 리스트
file_path = "/home/kimdu/hanse/src/ai_manipulation/ai_manipulation/datas/way_point_2.json"
with open(file_path, 'r') as json_file:
    my_list = json.load(json_file)
# 함수 호출
count_duplicates(my_list) 
```
